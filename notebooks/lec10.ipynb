```python
gov_col = ['countryname', 'ifs', 'year', 'system', 'finittrm', 
           'military', 'execme', 'execrlc', 'numgov', 'numopp', 'legelec', 'exelec']
minified_gov = df_gov[gov_col]
```

```python
# minified_gov['execme_num'] = 
labels, uniques = pd.factorize(minified_gov['execme']) 
minified_gov['execme_num'] = labels
```

```python
minified_gov.head(10)
```

```python
print(minified_gov.system.unique())
def is_system(row, value):
    if row['system'] == value:
        return 1.0
    return 0.0
```

```python
# Add new columns to properly distinguish system implemented
minified_gov['is_presidential'] = minified_gov.apply(lambda x: is_system(x, 'Presidential'), axis=1)
minified_gov['is_parliamentary'] = minified_gov.apply(lambda x: is_system(x, 'Parliamentary'), axis=1)
minified_gov['is_essembly_elected'] = minified_gov.apply(lambda x: is_system(x, 'Assembly-Elected President'), axis=1)
```

```python
minified_gov.head(10)
```

```python
print(minified_gov.execrlc.unique())
def is_party_orientation(row, value):
    if row['execrlc'] == value:
        return 1.0
    return 0.0

minified_gov['is_left_oriented'] = minified_gov.apply(lambda x: is_party_orientation(x, 'Left'), axis=1)
minified_gov['is_right_oriented'] = minified_gov.apply(lambda x: is_party_orientation(x, 'Right'), axis=1)
minified_gov['is_center_oriented'] = minified_gov.apply(lambda x: is_party_orientation(x, 'Center'), axis=1)
```

```python
minified_gov.head(10)
```

```python
minified_gov['numadmin'] = minified_gov.numgov - minified_gov.numopp
```
